## Supervised Learning: Classification (Amazon Example)
    - Problem: predict if a review is helpful or unhelpful
    - Data: 
        - Features: Star Rating, Length of Review
        - Labels: Helpful or Unhelpful

## Sample Table:
| Rating | Length | Helpfulness |
| :- | :- | :- |
| .6 | .7 | + |
| .2 | .2 | - |
| .8 | .2 | - |
| .2 | .9 | + |
| .6 | .4 | ? |

## Example Graph w/ Code:


```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn-whitegrid')

ratings_good = [.6, .2]

ratings_bad = [.2, .8]

lengths_good = [.7, .9]

lengths_bad = [.2, .2]

rating_unknown = [.6]

length_unknown = [.4]


x = np.linspace(0,.9, 100)

#This is an example, just showing the possibility for a linear boundary--not a derived one.
example_goal_linear_boundary_y = -.6*x+.8

plt.plot(x, example_goal_linear_boundary_y)

plt.plot(ratings_good, lengths_good, '+', color='green')

plt.plot(ratings_bad, lengths_bad, 'x', color='red')

plt.plot(rating_unknown, length_unknown, '*', color='black')

plt.text(.45, .83, r'*: Point of interest', fontsize=15)
plt.xlabel("Ratings")
plt.ylabel("Lengths")
```

## Why a Linear Decision Boundary?
### &ensp;&ensp;Over and Underfitting becomes a real problem.

### While the following graph has an imperfect decision boundary:

```python
ratings_good = [.6, .2, .7, .9]

ratings_bad = [.2, .8, .7, .1, .4]

lengths_good = [.7, .9, .4, .7]

lengths_bad = [.2, .2, .9, .8, .7]

rating_unknown = [.6]

length_unknown = [.4]


x = np.linspace(0,.9, 100)

#This is an example, just showing the possibility for a linear boundary--not a derived one.
example_goal_linear_boundary_y = -.6*x+.9

plt.plot(x, example_goal_linear_boundary_y)

plt.plot(ratings_good, lengths_good, '+', color='green')

plt.plot(ratings_bad, lengths_bad, 'x', color='red')

plt.plot(rating_unknown, length_unknown, '*', color='black')

plt.text(.45, .83, r'*: Point of interest', fontsize=15)
plt.xlabel("Ratings")
plt.ylabel("Lengths")
```

## It is still better than overfitting and coming up with something like this:

```python
ratings_good = [.6, .2, .7]

ratings_bad = [.2, .8, .7, .1, .4]

lengths_good = [.7, .9, .4]

lengths_bad = [.2, .2, .9, .8, .7]

rating_unknown = [.6]

length_unknown = [.4]

x = np.linspace(0,.9, 100)

example_goal_cubic_boundary_y = -2 + 22.92857*x - 47.85714*(x**2) + 28.57143*(x**3)

plt.plot(x, example_goal_cubic_boundary_y)

plt.plot(ratings_good, lengths_good, '+', color='green')

plt.plot(ratings_bad, lengths_bad, 'x', color='red')

plt.plot(rating_unknown, length_unknown, '*', color='black')

plt.text(.45, -1, r'*: Point of interest', fontsize=15)
plt.xlabel("Ratings")
plt.ylabel("Lengths")
```

## This can lead to poor prediction accuracy, while maintaining high training scores, due to its fitting exactly to the data.
